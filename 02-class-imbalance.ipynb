```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter
import seaborn as sns
import os
```

```python
# Load Preprocessed Data
DATA_PATH = "/kaggle/input/preprocessed-mammo-splits"  # Update if different
train = np.load(os.path.join(DATA_PATH, "train_data.npz"))
val = np.load(os.path.join(DATA_PATH, "val_data.npz"))
test = np.load(os.path.join(DATA_PATH, "test_data.npz"))

X_train, y_train = train["X"], train["y"]
X_val, y_val = val["X"], val["y"]
X_test, y_test = test["X"], test["y"]
```

```python
# Check Distribution
counter = Counter(y_train)
labels = list(counter.keys())
counts = list(counter.values())

plt.figure(figsize=(6, 4))
sns.barplot(x=labels, y=counts)
plt.xticks([0, 1], ["Benign", "Malignant"])
plt.title("Class Distribution in Training Data")
plt.ylabel("Count")
plt.show()
```

```python
# Compute Class Weights for Loss Functions
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = dict(zip(np.unique(y_train), class_weights))
print("\u2705 Computed Class Weights:", class_weight_dict)
```
